```python
import pandas as pd
import json
from nba_api.stats.endpoints import FranchiseHistory

# Disable row limit in pandas display
pd.set_option('display.max_rows', None)

# Get franchise history data
franchise_history = FranchiseHistory().get_data_frames()[0]

# Create a dictionary to store the data
franchises_data = {}

# Process each franchise
for i, row in franchise_history.iterrows():
    team_id = row['TEAM_ID']
    team_abbreviation = row.get('TEAM_ABBR', '')  # Some older franchises might not have abbreviations
    
    # Use team city and nickname if abbreviation is missing
    if not team_abbreviation:
        team_abbreviation = f"{row['TEAM_CITY']}_{row['TEAM_NAME']}"
    
    # Get start year
    start_year = row['START_YEAR']
    
    # Get end year (if available, otherwise set to current)
    end_year = row['END_YEAR'] if row['END_YEAR'] != 0 else "Present"
    
    # Store in dictionary
    franchises_data[i] = {
        "team_id": team_id,
        "team_city": row['TEAM_CITY'],
        "team_name": row['TEAM_NAME'],
        "abbreviation": team_abbreviation,
        "start_year": start_year,
        "end_year": end_year,
        "active": (end_year == "Present")
    }

# Save to JSON file
with open('nba_franchise_history.json', 'w') as json_file:
    json.dump(franchises_data, json_file, indent=4)

print("NBA franchise history data saved to 'nba_franchise_history.json'")
```

```python
import json

# Load the JSON file
with open('teams_data.json') as f:
    data = json.load(f)

unique_teams = set()

# Iterate through each team and their games
for team_id, team_info in data.items():
    for game in team_info.get("games", []):
        matchup = game.get("MATCHUP", "")
        parts = matchup.split()
        if len(parts) == 3:
            team1, _, team2 = parts
            unique_teams.add(team1)
            unique_teams.add(team2)

# Print unique team abbreviations
print(sorted(unique_teams))

```
